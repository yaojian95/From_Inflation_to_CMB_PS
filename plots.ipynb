## Generate maps from power spetra

$$\Large{a_{\ell m} = \sqrt{-log(A)C_{\ell}}*e^{2\pi i\theta}}$$

```python
import healpy as hp
plt.set_cmap('viridis')
```

```python
ells = data_ps[:,0]; nside = 128; lmax = 3*nside-1
cl = data_ps[:,1]/coeff
```

```python
ells = np.arange(2001)
```

```python
a = np.array((0,0))
cl_all = np.append(a, cl)
```

```python
alms_s = np.zeros_like(alms)
for ell in range(2, lmax+1):
    N = 0;
    for m in range(ell+1):
        A = np.random.random()
        theta = np.random.random()
        alms_s[N + ell - m] = np.sqrt(-np.log(A)*cl_all[ell])*np.exp(2*np.pi*1j*theta)

        n_i = (lmax + 1) - m  
        N += n_i
```

```python
map_sim = hp.alm2map(alms_s, nside = nside)
```

```python
def pick_alm(alms, lmax, ell):
    '''
    pick out the alms according to ell from the complete alms
    
    alms_ell = pick_alm(alms, 3*nside-1, 100)
    '''
    alms_ell = np.empty(ell+1, dtype = complex); N = 0
    alms_0 = np.zeros_like(alms)
    for m in range(ell+1):
        alms_ell[m] = alms[N + ell - m]
        alms_0[N + ell - m] = alms[N + ell - m]
        
        n_i = (lmax + 1) - m  
        N += n_i
    
    return alms_ell, alms_0
```
